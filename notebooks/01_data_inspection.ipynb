```python
import pandas as pd
from pathlib import Path

base_path = Path("data/raw")

files = {
    "contacts": base_path / "contacts.csv",
    "donors": base_path / "donors.csv",
    "events": base_path / "events.csv",
    "family": base_path / "family.csv",
    "giving": base_path / "giving.csv",
    "relationships": base_path / "relationships.csv",
}

dfs = {}
for name, path in files.items():
    dfs[name] = pd.read_csv(path)
    print(f"\n{name.upper()}")
    print("Shape:", dfs[name].shape)
    print("Columns:", list(dfs[name].columns))
    display(dfs[name].head())
```

```python
# set up working copies 
import pandas as pd
import numpy as np

donors = dfs["donors"].copy()
giving = dfs["giving"].copy()
contacts = dfs["contacts"].copy()
events = dfs["events"].copy()
relationships = dfs["relationships"].copy()
family = dfs["family"].copy()
```

```python
# fix date columns 
giving["Gift_Date"] = pd.to_datetime(giving["Gift_Date"], errors="coerce")
contacts["Contact_Date"] = pd.to_datetime(contacts["Contact_Date"], errors="coerce")
events["Attendance_Date"] = pd.to_datetime(events["Attendance_Date"], errors="coerce")

if "Last_Gift_Date" in donors.columns:
    donors["Last_Gift_Date"] = pd.to_datetime(donors["Last_Gift_Date"], errors="coerce")
```

```python
# build giving summary
giving_summary = giving.groupby("Donor_ID").agg(
    total_gift_amount=("Gift_Amount", "sum"),
    total_gift_count=("Gift_ID", "count"),
    avg_gift_amount=("Gift_Amount", "mean"),
    last_gift_date=("Gift_Date", "max"),
    first_gift_date=("Gift_Date", "min")
).reset_index()

giving_summary.head()
```

```python
# contacts summary
contacts_summary = contacts.groupby("Donor_ID").agg(
    total_contacts=("Contact_Report_ID", "count"),
    last_contact_date=("Contact_Date", "max"),
    positive_contacts=("Outcome_Category", lambda x: (x == "Positive").sum()),
    negative_contacts=("Outcome_Category", lambda x: (x == "Negative").sum())
).reset_index()

contacts_summary.head()
```

```python
# events summary
events_summary = events.groupby("Donor_ID").agg(
    total_events_attended=("Event_Name", "count"),
    last_event_date=("Attendance_Date", "max")
).reset_index()

events_summary.head() 
```

```python
# relationships summary
rel_out = relationships.groupby("Donor_ID_1").agg(
    total_relationships_out=("Donor_ID_2", "count"),
    avg_relationship_strength=("Relationship_Strength", "mean"),
    max_relationship_strength=("Relationship_Strength", "max")
).reset_index().rename(columns={"Donor_ID_1": "ID"})

rel_out.head()
```

```python
# choose donor columns
candidate_cols = [
    "ID", "Full_Name", "Gender", "donor_type", "Constituent_Type_2",
    "Primary_Manager", "Rating", "Prospect_Stage", "Lifetime_Giving",
    "Last_Gift", "Last_Gift_Designation", "Last_Gift_Date",
    "Engagement_Score", "engagement_score", "Will_Give_Again_Probability",
    "influence_score", "network_size", "net_work_size", "degree_centrality",
    "years_active", "recency_score", "monetary_score", "rfm_score",
    "region", "Professional_Background", "qualified_rating",
    "high_influence_role", "high_rating", "has_board_role"
]

donor_cols = [c for c in candidate_cols if c in donors.columns]
donor_cols
```

```python
# base table
prospect_pipeline = donors[donor_cols].copy()
prospect_pipeline.head()
```

```python
# merge giving
prospect_pipeline = prospect_pipeline.merge(
    giving_summary,
    left_on="ID",
    right_on="Donor_ID",
    how="left"
).drop(columns=["Donor_ID"], errors="ignore")

prospect_pipeline.head()
```

```python
# merge contacts
prospect_pipeline = prospect_pipeline.merge(
    contacts_summary,
    left_on="ID",
    right_on="Donor_ID",
    how="left"
).drop(columns=["Donor_ID"], errors="ignore")

prospect_pipeline.head()
```

```python
# merge events
prospect_pipeline = prospect_pipeline.merge(
    events_summary,
    left_on="ID",
    right_on="Donor_ID",
    how="left"
).drop(columns=["Donor_ID"], errors="ignore")

prospect_pipeline.head()
```

```python
# merge relationships
prospect_pipeline = prospect_pipeline.merge(
    rel_out,
    on="ID",
    how="left"
)

prospect_pipeline.head()
```

```python
# fill numeric nulls
numeric_cols = prospect_pipeline.select_dtypes(include=[np.number]).columns
prospect_pipeline[numeric_cols] = prospect_pipeline[numeric_cols].fillna(0)

print(prospect_pipeline.shape)
prospect_pipeline.head()
```

```python
# pick engagement column
if "engagement_score" in prospect_pipeline.columns:
    engagement_col = "engagement_score"
elif "Engagement_Score" in prospect_pipeline.columns:
    engagement_col = "Engagement_Score"
else:
    engagement_col = None

engagement_col
```

```python
# create BI flags
prospect_pipeline["has_contact"] = np.where(
    prospect_pipeline["total_contacts"] > 0, 1, 0
)

prospect_pipeline["has_event_attendance"] = np.where(
    prospect_pipeline["total_events_attended"] > 0, 1, 0
)

prospect_pipeline["is_high_value"] = np.where(
    prospect_pipeline["total_gift_amount"] >= 10000, 1, 0
)
```

```python
# engagement tiers
if engagement_col is not None:
    prospect_pipeline["engagement_tier"] = pd.qcut(
        prospect_pipeline[engagement_col].rank(method="first"),
        q=4,
        labels=["Low", "Moderate", "High", "Very High"]
    )
else:
    prospect_pipeline["engagement_tier"] = "Unknown"

prospect_pipeline[["ID", "Full_Name", engagement_col, "engagement_tier"]].head()
```

```python
# priority label
prospect_pipeline["portfolio_priority"] = np.select(
    [
        (prospect_pipeline["Will_Give_Again_Probability"] >= 0.75) &
        (prospect_pipeline["total_contacts"] <= 2),

        (prospect_pipeline["Will_Give_Again_Probability"] >= 0.75) &
        (prospect_pipeline["total_contacts"] > 2),

        (prospect_pipeline["total_gift_amount"] >= 10000) &
        (prospect_pipeline["total_contacts"] == 0)
    ],
    [
        "High Probability, Low Outreach",
        "High Probability, Engaged",
        "High Value, No Contact"
    ],
    default="Standard"
)

prospect_pipeline[
    [
        "ID", "Full_Name", "Prospect_Stage", "Primary_Manager",
        "total_gift_amount", "total_contacts", "portfolio_priority"
    ]
].head(10)
```

```python
# validation
print(prospect_pipeline.shape)
print(prospect_pipeline.columns.tolist())
```

```python
# save processed file
from pathlib import Path

output_path = Path("data/processed/prospect_pipeline.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)

prospect_pipeline.to_csv(output_path, index=False)
print(f"Saved to: {output_path.resolve()}")
```

## SQL Analysis 1: Prospect Pipeline by Stage

This query summarizes the donor pipeline by prospect stage.  
It shows how many donors are in each stage, how much giving is associated with each stage, and the average probability of giving again.

This is useful for understanding overall pipeline health and stage-level fundraising performance.

```python
# Prospect pipeline by stage
query = """
SELECT
    Prospect_Stage,
    COUNT(*) AS donor_count,
    ROUND(SUM(total_gift_amount), 2) AS total_giving,
    ROUND(AVG(total_gift_amount), 2) AS avg_giving,
    ROUND(AVG(Will_Give_Again_Probability), 3) AS avg_give_again_prob
FROM prospect_pipeline
GROUP BY Prospect_Stage
ORDER BY total_giving DESC;
"""

stage_summary = pd.read_sql_query(query, conn)
stage_summary
```

### Interpretation

The prospect pipeline is widest at the Identification stage, which contains the largest donor volume and the highest total giving. This suggests that early-stage prospecting represents the largest pool of future fundraising opportunity.

The Stewardship stage has the highest average giving per donor, indicating that retained and stewarded donors generate the strongest donor value on a per-donor basis.

Because average give-again probability is relatively similar across stages, prospect stage alone may not be the strongest differentiator of future giving likelihood.

## SQL Analysis 2: Portfolio Health by Primary Manager

This query evaluates portfolio performance for each primary manager.  
It measures portfolio size, total giving, average giving per donor, contact activity, and donor giving probability.

This helps identify which portfolios are strongest and where additional outreach may be needed.

```python
# Portfolio health by primary manager
query = """
SELECT
    Primary_Manager,
    COUNT(*) AS portfolio_size,
    ROUND(SUM(total_gift_amount), 2) AS total_portfolio_giving,
    ROUND(AVG(total_gift_amount), 2) AS avg_giving_per_donor,
    ROUND(AVG(total_contacts), 2) AS avg_contacts,
    ROUND(AVG(Will_Give_Again_Probability), 3) AS avg_give_again_prob
FROM prospect_pipeline
WHERE Primary_Manager IS NOT NULL
  AND Primary_Manager != 'None'
GROUP BY Primary_Manager
ORDER BY total_portfolio_giving DESC
LIMIT 15;
"""

manager_summary = pd.read_sql_query(query, conn)
manager_summary
```

### Interpretation

Portfolio sizes appear evenly distributed across managers, so the most meaningful differences are in total portfolio giving and average giving per donor rather than donor count alone.

This view helps identify which managers oversee the highest-value portfolios and where stronger fundraising outcomes may be associated with donor mix, portfolio quality, or relationship management effectiveness.

## SQL Analysis 3: Donor Priority Segments

This query groups donors by the business-friendly `portfolio_priority` label created during data preparation.

It helps identify:
- high-probability donors with low outreach
- high-probability donors who are already engaged
- high-value donors with no contact
- standard donors

This type of segmentation can support outreach planning and prospect management.

```python
# Donor counts and giving by priority segment
query = """
SELECT
    portfolio_priority,
    COUNT(*) AS donor_count,
    ROUND(SUM(total_gift_amount), 2) AS total_giving,
    ROUND(AVG(total_gift_amount), 2) AS avg_giving
FROM prospect_pipeline
GROUP BY portfolio_priority
ORDER BY total_giving DESC;
"""

priority_summary = pd.read_sql_query(query, conn)
priority_summary
```

### Interpretation

The High Probability, Low Outreach segment stands out as the strongest immediate opportunity for advancement action. This group combines high projected giving likelihood with limited contact history, suggesting that targeted outreach could unlock near-term fundraising gains.

The High Value, No Contact segment is much smaller in volume, but its very high average giving makes it especially important for focused follow-up.

## SQL Analysis 4: High-Potential, Under-Engaged Donors

This query identifies donors with a high probability of giving again but very limited contact activity.

These records are especially useful because they may represent strong prospects who are not receiving enough outreach attention.

```python
# High-potential donors with limited outreach
query = """
SELECT
    ID,
    Full_Name,
    Prospect_Stage,
    total_gift_amount,
    total_contacts,
    total_events_attended,
    Will_Give_Again_Probability,
    influence_score,
    engagement_tier
FROM prospect_pipeline
WHERE Will_Give_Again_Probability >= 0.80
  AND total_contacts <= 1
ORDER BY Will_Give_Again_Probability DESC, total_gift_amount DESC
LIMIT 20;
"""

under_engaged = pd.read_sql_query(query, conn)
under_engaged
```

### Interpretation

This list highlights specific donors who combine:
- strong projected likelihood of giving again
- limited outreach activity
- existing giving history

These donors would be strong candidates for targeted cultivation, personalized follow-up, or portfolio review by advancement leadership.

## SQL Analysis 5: High-Potential Donors with Giving History and Limited Outreach

This query refines the under-engaged donor list by focusing on prospects who have both:
- a high probability of giving again
- limited contact activity
- some prior giving history

This creates a more actionable prospecting list for advancement teams.

```python
# High-potential donors with giving history and limited outreach
query = """
SELECT
    ID,
    Full_Name,
    Prospect_Stage,
    total_gift_amount,
    total_contacts,
    total_events_attended,
    Will_Give_Again_Probability,
    influence_score,
    engagement_tier,
    portfolio_priority
FROM prospect_pipeline
WHERE Will_Give_Again_Probability >= 0.80
  AND total_contacts <= 1
  AND total_gift_amount > 1000
ORDER BY Will_Give_Again_Probability DESC, total_gift_amount DESC
LIMIT 20;
"""

target_prospects = pd.read_sql_query(query, conn)
target_prospects
```

### Interpretation

This query produces a more practical prospecting list by narrowing the focus to donors with both prior giving history and high projected likelihood of giving again, but little recent outreach.

That makes this one of the most actionable outputs in the project, since it highlights prospects who may be good candidates for immediate cultivation or portfolio review.

## Create Power BI-Ready Dataset

This step standardizes numeric fields before loading the dataset into Power BI.  
It prevents import issues caused by mixed data types in CSV columns.

```python
powerbi_df = prospect_pipeline.copy()
```

```python
numeric_cols = [
    "ID",
    "Lifetime_Giving",
    "engagement_score",
    "Will_Give_Again_Probability",
    "influence_score",
    "network_size",
    "degree_centrality",
    "years_active",
    "recency_score",
    "monetary_score",
    "rfm_score",
    "qualified_rating",
    "total_gift_amount",
    "total_gift_count",
    "avg_gift_amount",
    "total_contacts",
    "positive_contacts",
    "negative_contacts",
    "total_events_attended",
    "total_relationships_out",
    "avg_relationship_strength",
    "max_relationship_strength",
    "has_contact",
    "has_event_attendance",
    "is_high_value"
]

for col in numeric_cols:
    if col in powerbi_df.columns:
        powerbi_df[col] = pd.to_numeric(powerbi_df[col], errors="coerce")
```

```python
date_cols = ["first_gift_date", "last_gift_date", "last_contact_date", "last_event_date"]

for col in date_cols:
    if col in powerbi_df.columns:
        powerbi_df[col] = pd.to_datetime(powerbi_df[col], errors="coerce")
```
